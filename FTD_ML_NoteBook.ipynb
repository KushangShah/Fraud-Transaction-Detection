```python
# Analyze and handle outliers in 'TX_AMOUNT'
# Using capping at the 99th percentile as an example
upper_limit = df['TX_AMOUNT'].quantile(0.99)
df['TX_AMOUNT'] = np.where(df['TX_AMOUNT'] > upper_limit, upper_limit, df['TX_AMOUNT'])
```

```python
df.describe()
```

Slight change in mean, max and std in `TX_AMOUNT`.

```python
# let's create new features as per the EDA done before.


# Flag if amount > 220
df['AMOUNT_OVER_220'] = (df['TX_AMOUNT'] > 220).astype(int)

# Average spend per customer and ratio
df['CUSTOMER_MEAN_SPEND'] = df.groupby('CUSTOMER_ID')['TX_AMOUNT'].transform('mean')
df['SPENDING_RATIO'] = df['TX_AMOUNT'] / (df['CUSTOMER_MEAN_SPEND'] + 1e-9)

# Terminal Fraud Count - last 28 days
df['TERMINAL_FRAUD_COUNT'] = 0
df = df.sort_values(['TERMINAL_ID', 'TX_DATETIME'])

# group by terminal
terminal_groups = df.groupby('TERMINAL_ID')
counts = []

# Calculate rolling fraud count for each terminal
for terminal, group in terminal_groups:
    q = deque()
    group_counts = []
    
    # iterate through each transaction in the group
    for i, row in group.iterrows():
        current_time = row['TX_DATETIME']
        while q and (current_time - q[0]) > timedelta(days=28):
            q.popleft()
        group_counts.append(len(q))
        
        # Add current fraud to the queue if applicable
        if row.get('TX_FRAUD', 0) == 1:
            q.append(current_time)
    counts.extend(group_counts)

# create the new feature
df['TERMINAL_FRAUD_COUNT'] = counts

    
```

### Hypothetical Statement - 1: Fraudulent transaction have higher transaction amount.

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

- **Null Hypothesis (H‚ÇÄ)**: The average transaction amount is the same for fraud and non fraud transactions.<br>
i.e., `H‚ÇÄ‚Äã: Œºfraud ‚Äã= Œºlegit`.<br>
‚ÄãŒºfraud = The true mean of fraudulent data.<br>
Œºlegit = The true mean of legitimate data.<br>

- **Alternative Hypothesis (H‚ÇÅ)**: Fradulant transactions have higher average transaction than legitimate transaction.<br>
i.e., `H‚ÇÅ: Œºfraud‚Äã > Œºlegit`<br>

#### 2. Perform an appropriate statistical test.

```python
# Perform Statistical Test to obtain P-Value
fraud_amounts = df[df['TX_FRAUD'] == 1]['TX_AMOUNT']
legit_amounts = df[df['TX_FRAUD'] == 0]['TX_AMOUNT']

t_stat, p_value = stats.ttest_ind(fraud_amounts, legit_amounts, equal_var=False, alternative='greater')

# Check output
print(f"\nT-Statistic from t-test: {t_stat}")
print(f"P-Value from t-test: {p_value}")

# Interpret the P-Value
alpha = 0.05
if p_value < alpha:
    print("\n‚úÖ Reject H‚ÇÄ ‚Äî Fraudulent transactions have significantly higher amounts.")
else:
    print("\n‚ùå Fail to reject H‚ÇÄ ‚Äî No significant difference found.")
```

##### Which statistical test have you done to obtain P-Value?

I have performed t-test statistic to obtain P-Value.

##### Why did you choose the specific statistical test?

This test case help in determining weather the mean of two different group are statistically significantly different from each other. 

### Hypothetical Statement - 2: Higher spending ratio indicats fraud.

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

- **Null Hypothesis (H‚ÇÄ)**: Mean spending ratio `SPENDING_RATIO` is the same for fraud  and non-fraud.
- **Alternative Hypothesis (H‚ÇÅ)**: Fraudulent transactions have higher mean spending ratio.

#### 2. Perform an appropriate statistical test.

```python
# Perform Statistical Test to obtain P-Value
fraud_amounts = df.loc[df['TX_FRAUD'] == 1, "SPENDING_RATIO"]
legit_amounts = df.loc[df['TX_FRAUD'] == 0, "SPENDING_RATIO"]

# perform test
t_test, p_value = stats.ttest_ind(fraud_amounts, legit_amounts, equal_var=False, alternative='greater')

# print output
print(f"\nT-Statistic from t-test: {t_test}")
print(f"P-Value from t-test: {p_value}")

alpha = 0.05

# interpret p-value
if p_value < alpha:
    print("\n‚úÖ Reject H‚ÇÄ ‚Äî Fraudulent transactions have significantly higher spending ratios.")
else:
    print("\n‚ùå Fail to reject H‚ÇÄ ‚Äî No significant difference found.")
```

### Hypothetical Statement - 3: Terminals with more history have higher fraud probability in the new transaction.

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

- **Null Hypothesis (H‚ÇÄ)**:Fraud rate is independent of terminal's historical fraud count.

- **Alternative Hypothesis (H‚ÇÅ)**: Terminals with high fraud count have a greater fraud rate in current transaction.

#### 2. Perform an appropriate statistical test.

```python
# Create categorical bins for terminal fraud history
df['TERMINAL_BIN'] = pd.cut(df['TERMINAL_FRAUD_COUNT'], bins=[0,1,5,10,50,1000], labels=['0','1-5','6-10','11-50','>50'])

# Contingency table
contingency = pd.crosstab(df['TERMINAL_BIN'], df['TX_FRAUD'])
print(contingency)

# Chi-square test
chi2, p, dof, expected = stats.chi2_contingency(contingency)
print("\nChi-square statistic:", chi2)
print("P-value:", p)

# Interpret the P-Value
if p < 0.05:
    print("\n‚úÖ Reject H‚ÇÄ ‚Äî Fraud rate depends on terminal fraud history.")
else:
    print("\n‚ùå Fail to reject H‚ÇÄ ‚Äî No significant relationship found.")
```

## ***7. ML Model Implementation***

### 8. Data Splitting

```python
def balance_dataset(train_df, target_col='TX_FRAUD'):
    """Downsamples non-fraud transactions to balance the dataset."""
    legit = train_df[train_df[target_col] == 0]
    fraud = train_df[train_df[target_col] == 1]

    legit_down = resample(
        legit,
        replace=False,
        n_samples=min(len(legit), len(fraud) * 5),  # keep ratio ~5:1
        random_state=42
    )

    balanced = pd.concat([fraud, legit_down])
    balanced = balanced.sample(frac=1, random_state=42).reset_index(drop=True)

    print(f"‚úÖ Balanced dataset created with {len(balanced)} rows")
    print(f"Fraud ratio: {balanced[target_col].mean():.2%}")
    return balanced

train_balanced = balance_dataset(df)
```

```python
# Define feature and target

features = ['TX_AMOUNT', 'AMOUNT_OVER_220', 'TERMINAL_FRAUD_COUNT', 'SPENDING_RATIO']
target = 'TX_FRAUD'

X_train = train_balanced[features]
y_train = train_balanced[target]

X_test = df[features]
y_test = df[target]

print("Data ready for model training.")
print(f"Training rows: {len(X_train):,}, Testing rows: {len(X_test):,}")
```

### ML Models

```python
# TRAIN AND COMPARE MULTIPLE MODELS

def train_and_compare_models(X_train, y_train, X_test, y_test):
    # Dictionary to store different machine learning models
    models = {
        "Logistic Regression": LogisticRegression(max_iter=500, solver='liblinear'),
        "Decision Tree": DecisionTreeClassifier(random_state=42),
        "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1),
        "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
        "LightGBM": LGBMClassifier(random_state=42)
    }

    # List to hold results for each model
    results = []
    
    # Iterate through each model in the dictionary
    for name, model in models.items():
        print(f"\nüöÄ Training {name}...")  # Indicate which model is currently being trained
        
        # Fit the model to the training data
        model.fit(X_train, y_train)
        
        # Predict the classes for the test data
        y_pred = model.predict(X_test)
        
        # Predict the probabilities for the positive class
        y_prob = model.predict_proba(X_test)[:, 1]
        
        # Calculate evaluation metrics
        precision = precision_score(y_test, y_pred, zero_division=0)  # Precision of the predictions
        recall = recall_score(y_test, y_pred, zero_division=0)        # Recall of the predictions
        f1 = f1_score(y_test, y_pred, zero_division=0)                # F1 score of the predictions
        roc = roc_auc_score(y_test, y_prob)                            # ROC AUC score of the predictions

        # Append the results for the current model to the results list
        results.append({
            "Model": name,
            "Precision": precision,
            "Recall": recall,
            "F1-Score": f1,
            "ROC-AUC": roc
        })

    # Convert results to a DataFrame and sort by ROC AUC score in descending order
    results_df = pd.DataFrame(results).sort_values(by="ROC-AUC", ascending=False)
    
    # Display the results DataFrame
    display(results_df)
    
    # Return the results DataFrame for further analysis or reference
    return results_df

# Call the function with the training and testing datasets and store the results
results_df = train_and_compare_models(X_train, y_train, X_test, y_test)
```
